```python
# Write a Python program to triple all numbers of a given list of integers. Use Python map
lst = [5,12,17,45,86]
op_lst = list(map(lambda x : x * 3,lst))
print(op_lst)
```

```python
# Write a Python program to add three given lists using Python map and lambda
lst1 = [28,10,4,31]
lst2 = [112,103,41,1]
lst3 = [248,7,5,74]

op = list(map(lambda x,y,z : x+y+z,lst1,lst2,lst3))
print(op)
```

```python
# Write a Python program to add two given lists and find the difference between lists

"""
sample list
nums1 = [6, 5, 3, 9]
nums2 = [0, 1, 7, 7]


output  - [(6, 6), (6, 4), (10, -4), (16, 2)]

"""

nums1 = [6, 5, 3, 9]
nums2 = [0, 1, 7, 7]
counts = -1
def adds(x):
  global counts
  counts += 1
  return tuple( (x + nums2[counts],x - nums2[counts]))

op = list(map(adds,nums1))
print(op)
```

```python
# Write a Python program to convert all the characters in uppercase and lowercase and eliminate duplicate letters from a given sequence
ip = ['Hello','Test','Wow','Mirror']
def converts(x):
  op = []
  for i in x:
    if op.count(i.lower()) == 0 and op.count(i.upper()) == 0:
      op.append(i)
  return "".join(op)
uppers = list(map(lambda x : x.upper(),ip))
lowers = list(map(lambda x : x.lower(),ip))
dup = list(map(converts,ip))
print(uppers)
print(lowers)
print(dup)
```

```python
# Filter out the fruits that start with A
fruits = ['Mango','Apple','Apricot','Grapes','Orange','Avacado']
op_fruits = list(filter(lambda x : x.upper().startswith('A'),fruits))
print(op_fruits) 
```

```python
# Filter out the positve number from a list
nums = [10,-2,-3,7,100,-70]
def positive(num):
  if num > 0:
    return True
  return False
res = list(filter(positive,nums))
print(res)
```

```python
# filter out the multiples of 3 and 5 from a list
nums = [10,15,21,30,60,100]
op = list(filter(lambda x : x%3 == 0 and x%5 ==0,nums))
print(op)
```

```python
# find the max string using reuduce function
from functools import reduce
ip = ['Test','Hello','Welcome','Hi','Yes']
op = list(reduce(lambda x,y : x if len(x) > len(y) else y,ip))
print("".join(op))
```

```python
# find the mean , max and min of a given list using reduce function
from functools import reduce
nums = [10,15,21,30,60,100]
def maxs(x,y):
  if x > y:
    return x
  else:
    return y
def mins(x,y):
  if x < y:
    return x
  else:
    return y
avg = reduce(lambda x,y : x+y,nums) / len(nums)
max = reduce(maxs,nums)
min = reduce(mins,nums)
print(avg)
print(max)
print(min)
```
